```python
from transformers import AutoModelForCausalLM, AutoTokenizer, Trainer, TrainingArguments
from datasets import Dataset
import json
import torch

# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load JSON data
file_path = "llama_friendly.json"  # Update with your JSON file path
with open(file_path, "r") as f:
    data = json.load(f)

# Create Hugging Face Dataset
dataset = Dataset.from_dict({
    "instruction": [item["instruction"] for item in data],
    "response": [item["response"] for item in data]
})

# Load tokenizer and model (switching to a smaller model like 'distilgpt2')
model_name = "distilgpt2"  # A smaller version of GPT-2
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Set padding token as the same as EOS token (common for GPT models)
tokenizer.pad_token = tokenizer.eos_token  # Use EOS token as padding token

model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float32, device_map="auto")

# Move model to the available device (GPU if available)
model.to(device)

# Tokenization function with much smaller max_length (64)
def tokenize_function(examples):
    text = [f"Instruction: {ins}\nResponse: {resp}" for ins, resp in zip(examples["instruction"], examples["response"])]
    encoding = tokenizer(text, padding="max_length", truncation=True, max_length=64)  # Reduced max_length to 64
    encoding["labels"] = encoding["input_ids"]  # For causal language modeling, labels = input_ids
    return encoding

# Tokenize the entire dataset
tokenized_datasets = dataset.map(tokenize_function, batched=True)

# Split the dataset into training (no evaluation here)
train_dataset = tokenized_datasets

# Set training arguments with even smaller batch size and more gradient accumulation
training_args = TrainingArguments(
    output_dir="./results",  # Directory to save the model
    evaluation_strategy="no",  # No evaluation during training
    per_device_train_batch_size=1,  # Use batch size 1
    gradient_accumulation_steps=8,  # Accumulate gradients to simulate a larger batch size
    logging_dir="./logs",  # Directory for logging
    logging_steps=500,  # Log every 500 steps
    save_steps=1000,  # Save model every 1000 steps
    save_total_limit=2,  # Keep only the last 2 saved models to save space
    fp16=False,  # Disable FP16 to avoid mixed precision issues
)

# Initialize Trainer
trainer = Trainer(
    model=model,  # Pretrained model
    args=training_args,  # Training arguments
    train_dataset=train_dataset,  # Training dataset
)

# Start training
trainer.train()

# Save the fine-tuned model
trainer.save_model("./fine_tuned_model")

# Clear GPU memory to prevent memory buildup
torch.cuda.empty_cache()
```

```python
# Example prompt
prompt = "Unable to access email"

# Tokenize the input prompt
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

# Set attention mask and pad token id if necessary
inputs["attention_mask"] = inputs["attention_mask"] if "attention_mask" in inputs else torch.ones(inputs["input_ids"].shape, device=model.device)
inputs["pad_token_id"] = tokenizer.eos_token_id  # Setting pad token ID to eos token ID

# Generate output (e.g., 50 tokens)
outputs = model.generate(inputs["input_ids"], attention_mask=inputs["attention_mask"], pad_token_id=inputs["pad_token_id"], max_length=50, num_return_sequences=1)

# Decode the output into text
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("Model response:", generated_text)

```
